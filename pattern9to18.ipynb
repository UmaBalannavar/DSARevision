```python
#pattern 11
n=5
for i in range(n):
  if i%2==0: flip=1
  else: flip=0
  for j in range(i+1):
    print(flip,end="")
    flip=1-flip
  print()
```

```python
#pattern 11 or
n=5
for i in range(n):
  flip=1 if i%2==0 else 0
  for j in range(i+1):
    print(flip,end="")
    flip=1-flip
  print()
```

```python
#pattern 12
n=5
for i in range(n):
  count=1 # reset to 1 everytime
  for j in range(i+1):
    print(count,end="")
    count=count+1
  for j in  range(2*(n-i-1)):
    print(" ",end="")
  for j in range(i+1):
    count=count-1
    print(count,end="")
  print()
```

```python
#pattern 13
n=5
count=1
for i in range(n):
  for j in range(i+1):
    print(count,end="")
    count=count+1
  print()
```

```python
#pattern 17
for i in range(n):
  alp=65
  for j in range(n-i-1):
    print(" ",end="")
  for j in range(i+1):
    print(chr(alp),end="")
    alp=alp+1
  alp=alp-1
  for j in range(i):
    alp=alp-1
    print(chr(alp),end="")
  print()
```

```python
n=5
for i in range(n):
  print((n-i-1)*" ",end="")
  for j in range(i+1):
    print(chr(65+j),end="")
  for j in range(i-1,-1,-1):
    print(chr(65+j),end="")
  print()
```
